```python
import numpy as np
import pandas as pd
from apyori import apriori
```

```python
data = pd.read_csv('Pizza Data.csv', header=None)
data.head()
```

```python
df = pd.DataFrame(data)
df[0]=df.index
df
```

```python
df= df[1].str.split(';', expand=True)
df.to_csv("pizza.csv")
```

```python
pizza_toppings = []

for i in range(0, len(df)):
    pizza_toppings.append([str(df.values[i,j]) for j in range(0, len(df.columns))])
```

```python
#play with these numbers and learn about apriori
apriori_rule = apriori(pizza_toppings, 
                            min_support=0.05, 
                            min_confidence=0.5, 
                            min_lift=2, 
                            min_length=2)

apriori_results = list(apriori_rule)
```

```python
apriori_results
```

```python
#allison's code to extract information from the apriori algorithm and put it in a cleaner looking table
def inspect(output):
    I_val      = [tuple(result[2][0][0])[0] for result in output]
    j_val      = [tuple(result[2][0][1])[0] for result in output]
    support    = [result[1] for result in output]
    confidence = [result[2][0][2] for result in output]

    return list(zip(I_val, j_val, support, confidence))
  
```

```python
association = pd.DataFrame(inspect(apriori_results), columns = ['I', 'j', 'support', 'confidence'])
association = association[association['j'] != 'None']
association = association[association['I'] != 'None']

association
```
